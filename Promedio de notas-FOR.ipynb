```python
n = int(input("Ingrese la cantidad de notas: "))
suma = 0

for i in range(1, n + 1):
    nota = float(input(f"Ingrese la nota {i}: "))
    suma += nota

promedio = suma / n
print(f"Promedio final: {promedio}")
```
