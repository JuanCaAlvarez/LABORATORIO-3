```python
monto = float(input("Ingrese el monto de la compra; "))

if monto > 500:
  descuento = 0.20
elif monto > 200:
  descuento = 0.10
else:
  descuento = 0.05

total = monto - (monto * descuento)
print (f"Descuento aplicado: {descuento * 100}%")
print (f"Total a pagar: {total}")
```
